## Introduccion

El siguiente proyectoi es un recomendador d peliculas sencillo, basado en MovieLens con 100K de datos para entreno

# Primeros pasos

Importaci√≥n e las librerias b√°sicas de Panas y N

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
import ipywidgets as widgets
from IPython.display import display

import warnings
warnings.filterwarnings("ignore")
```

## Carga de Datos

Se hace la carga de los datasets descargados desde Kaggle


```python
# Cargar datos
atings = pd.read_csv("ml-100k/u.data", sep="\t", names=["user_id", "movie_id", "rating", "timestamp"])
movies = pd.read_csv("ml-100k/u.item", sep="|", encoding="latin-1", header=None, usecols=[0, 1], names=["movie_id", "title"])
df = pd.merge(ratings, movies, on="movie_id")
df["liked"] = (df["rating"] >= 4).astype(int)
```

## Crear la matriz de usuarios/pel√≠culas

(pivot_table) reorganiza el Dataframe original (dF) para crear una nueva tabla 


```python
# Datos base
peliculas = sorted(df["title"].unique())
movie_titles = df[["movie_id", "title"]].drop_duplicates().set_index("movie_id")["title"]
title_to_id = {title: mid for mid, title in movie_titles.items()}
```

## Funci√≥n de similitud personalizada

Se genera columna binaria seg√∫n el m√≠nimo rating dado, una matriz de ratings actualizada y calculamos la correlaci√≥n 

```python
# Configurar modelo
scaler = StandardScaler()
clf = RandomForestClassifier()
# Entrenamiento inicial
X = df[["user_id", "movie_id"]]
y = df["liked"]
X_scaled = scaler.fit_transform(X)
clf.fit(X_scaled, y)
```

 ## recomendaciones por correlaci√≥n
 se genera la funcion que correlaciones los datos del codigo 

```python
def get_similar_movies(movie_name, min_ratings=50, min_rating_value=4):
    movie_matrix = df.pivot_table(index="user_id", columns="title", values="rating")
    counts = df.groupby("title")["rating"].count()
    corr_scores = movie_matrix.corrwith(movie_matrix[movie_name])
    corr_df = pd.DataFrame(corr_scores, columns=["correlation"]) 
    corr_df = corr_df.join(counts.rename("count"))
    means = df.groupby("title")["rating"].mean()
    corr_df = corr_df.join(means.rename("mean_rating"))
    recs = corr_df[(corr_df["count"] >= min_ratings) & (corr_df["mean_rating"] >= min_rating_value)]
    return recs.sort_values("correlation", ascending=False)
```

## Widgets para recomendaciones basadas en correlaci√≥n

```python
selector_pelicula = widgets.Dropdown(options=peliculas, description='Pel√≠cula:', layout=widgets.Layout(width='80%'))
selector_rating = widgets.IntSlider(value=4, min=1, max=5, step=1, description='Rating m√≠nimo:', layout=widgets.Layout(width='50%'))
boton_similares = widgets.Button(description="Ver similares üé¨")
output_similares = widgets.Output()

def on_similares_click(b):
    with output_similares:
        output_similares.clear_output()
        pelicula = selector_pelicula.value
        rating_min = selector_rating.value
        resultado = get_similar_movies(pelicula, min_ratings=50, min_rating_value=rating_min).head(10)
        display(widgets.HTML(f"<h4>üéûÔ∏è Pel√≠culas similares a: <em>{pelicula}</em> con rating m√≠nimo {rating_min}‚≠ê</h4>"))
        display(resultado)

boton_similares.on_click(on_similares_click)
```

## Widgets para enviar rating del usuario y reentrenar modelo

```python
new_user_id = df["user_id"].max() + 1
rating_movie = widgets.Dropdown(options=peliculas, description='Tu pel√≠cula:', layout=widgets.Layout(width='80%'))
rating_value = widgets.IntSlider(value=4, min=1, max=5, step=1, description='Tu rating:', layout=widgets.Layout(width='50%'))
rate_button = widgets.Button(description="Enviar rating ‚ù§Ô∏è")
rate_output = widgets.Output()

def on_rate_click(b):
    with rate_output:
        rate_output.clear_output()
        pelicula = rating_movie.value
        movie_id = title_to_id[pelicula]
        timestamp = int(pd.Timestamp.now().timestamp())
        # Agregar nueva valoraci√≥n
        df.loc[len(df)] = [new_user_id, movie_id, rating_value.value, timestamp, pelicula, int(rating_value.value >= 4)]
        # Reentrenar modelo
        X = df[["user_id", "movie_id"]]
        y = df["liked"]
        X_scaled = scaler.fit_transform(X)
        clf.fit(X_scaled, y)
        print(f"Rating agregado: {pelicula} -> {rating_value.value}. Modelo actualizado.")

rate_button.on_click(on_rate_click)

# Mostrar widgets
display(widgets.VBox([selector_pelicula, selector_rating, boton_similares, output_similares]))
display(widgets.VBox([rating_movie, rating_value, rate_button, rate_output]))
```
