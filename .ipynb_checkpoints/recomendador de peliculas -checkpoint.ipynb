## Introduccion

El siguiente proyectoi es un recomendador d peliculas sencillo, basado en MovieLens con 100K de datos para entreno

# Primeros pasos

Importación e las librerias básicas de Panas y N

```python
import pandas as pd
import numpy as np
```

## Carga de Datos

Se hace la carga de los datasets descargados desde Kaggle


```python
# Ratings
ratings = pd.read_csv("ml-100k/u.data", sep="\t", names=["user_id", "movie_id", "rating", "timestamp"])

# Movie info
movies = pd.read_csv("ml-100k/u.item", sep="|", encoding="latin-1", header=None, usecols=[0, 1], names=["movie_id", "title"])
```

## Relacion de datos

Se correlacionan los datos, usando la función Merge que se encarga de relacionar un identificador con los datos de otro conjunto de datos


```python
df = pd.merge(ratings, movies, on="movie_id")
df.head()
```

## Pivote de datos

Se hace una relacion de los datos para describir qué puntuación le da cada usuario a cada titulo

```python
movie_matrix = df.pivot_table(index="user_id", columns="title", values="rating")
movie_matrix.head()
```

## Definición de función para recomendar (ML)

Se define una función que tendrá en cuenta datos de peliculas similares y su punutación para recomendarnos

```python
def get_similar_movies(movie_name, min_ratings=50):
    movie_ratings = movie_matrix[movie_name]
    similar_movies = movie_matrix.corrwith(movie_ratings)
    
    corr_df = pd.DataFrame(similar_movies, columns=["correlation"])
    corr_df.dropna(inplace=True)

    rating_counts = df.groupby("title")["rating"].count()
    corr_df = corr_df.join(rating_counts)

    return corr_df[corr_df["rating"] >= min_ratings].sort_values("correlation", ascending=False)
```

## Ejecución de la recomendación

Acontinuación llamamos a la fución co un dato para ejecutar las 10 mejores peliculas parecidas.

```python
get_similar_movies("Star Wars (1977)").head(10)
```

```python
get_similar_movies("101 Dalmatians (1996)").head(10)
```
